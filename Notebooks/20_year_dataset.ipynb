```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_excel('/content/drive/MyDrive/Jel jeevan mission/Yearwise dataset.xlsx')
data
```

```python
df = pd.DataFrame(data)
df = df.dropna()
df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Min-Max scale the selected columns
scaler = MinMaxScaler()
df[['Roads', 'Houses', 'Water supply']] = scaler.fit_transform(df[['Roads', 'Houses', 'Water supply']])

print(df)
```

```python
df.columns
```

```python
X = df.drop(['Year', 'IMR'], axis = 1)  # Predictor variables
y = df['IMR']  # Target variable
```

```python
model = LinearRegression()
model.fit(X, y)
coefficients2 = dict(zip(X.columns, model.coef_))
print("Weightage of changes in each attribute on changes in the target variable over 4 years:")
print(coefficients2)
```

Hereâ€™s a breakdown of what these results imply:

MMR (Maternal Mortality Rate):

Coefficient: 0.0318
Interpretation: A unit increase in MMR is associated with a small positive increase in the target variable (IMR, Infant Mortality Rate) by approximately 0.0318 units, holding all other variables constant. The effect of MMR on IMR is relatively minor.
Roads:

Coefficient: -13.026
Interpretation: A unit increase in the 'Roads' attribute is associated with a significant decrease in IMR by approximately 13.026 units, holding all other variables constant. This suggests that improvements in road infrastructure have a substantial positive impact on reducing IMR.
Houses:

Coefficient: -13.140
Interpretation: A unit increase in the 'Houses' attribute is associated with a significant decrease in IMR by approximately 13.140 units, holding all other variables constant. This indicates that better housing conditions significantly contribute to reducing IMR.
Water supply:

Coefficient: 0.1919
Interpretation: A unit increase in the 'Water supply' attribute is associated with a small positive increase in IMR by approximately 0.1919 units, holding all other variables constant. This suggests a minor effect of water supply improvements on IMR, though this result might need further investigation, as it appears counterintuitive.

```python
Karnataka_dataset = pd.read_excel('/content/drive/MyDrive/PMGSY/Karnataka 20 year dataset.xlsx')
Karnataka_dataset
```

```python
scaler = MinMaxScaler()
Karnataka_dataset[['Roads', 'Houses', 'JSY']] = scaler.fit_transform(Karnataka_dataset[['Roads', 'Houses', 'JSY']])
Karnataka_dataset
```

```python
Karnataka_X = Karnataka_dataset.drop(['Year', 'IMR'], axis = 1)  # Predictor variables
y = Karnataka_dataset['IMR']  # Target variable
model = LinearRegression()
model.fit(Karnataka_X, y)
coefficients2 = dict(zip(Karnataka_X.columns, model.coef_))
print("Weightage of changes in each attribute on changes in the target variable over 4 years:")

for x in coefficients2:
  print(x, "-> ",coefficients2[x])
```

The coefficients provide insight into how each predictor variable influences the target variable:

1. MMR: The increase in maternal mortality rate slightly increases infant mortality rate, suggesting a small direct relationship.
2. Roads: Improvement in road infrastructure has a moderate positive relationship with the target variable, which might indicate increased access or usage affecting IMR.
2. Houses: Improvement in housing conditions has a substantial positive relationship with the target variable, indicating significant influence.
3. JSY: The negative coefficient indicates that the Janani Suraksha Yojana program effectively reduces the target variable (IMR), showing its importance in maternal and infant health interventions.

```python
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

df = pd.DataFrame(Karnataka_dataset)

# Split the data into training (first 15 years) and test (last 5 years) sets
train = df.iloc[:15]
test = df.iloc[15:]

# Prepare the data
X_train = train[['Roads', 'Houses', 'JSY', 'MMR']]
y_train = train['IMR']
X_test = test[['Roads', 'Houses', 'JSY', 'MMR']]
y_test = test['IMR']

# Fit the multiple linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions for the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Display the coefficients
coefficients = pd.DataFrame(model.coef_, X_train.columns, columns=['Coefficient'])
print(coefficients)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(train['Year'], y_train, label='Training Data')
plt.plot(test['Year'], y_test, label='Actual Data', color='red')
plt.plot(test['Year'], y_pred, label='Predicted Data', color='green')
plt.xlabel('Year')
plt.ylabel('IMR')
plt.title('IMR Prediction')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

df = pd.DataFrame(Karnataka_dataset)

# Split the data into training (first 15 years) and test (last 5 years) sets
train = df.iloc[:15]
test = df.iloc[15:]

# Prepare the data
X_train = train[['Roads', 'Houses', 'JSY']]
y_train = train['IMR']
X_test = test[['Roads', 'Houses', 'JSY']]
y_test = test['IMR']

# Polynomial Features
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Fit the multiple linear regression model with polynomial features
model = LinearRegression()
model.fit(X_train_poly, y_train)

# Make predictions for the test set
y_pred = model.predict(X_test_poly)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Display the coefficients
coefficients = pd.DataFrame(model.coef_, poly.get_feature_names_out(X_train.columns), columns=['Coefficient'])
print(coefficients)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(train['Year'], y_train, label='Training Data')
plt.plot(test['Year'], y_test, label='Actual Data', color='red')
plt.plot(test['Year'], y_pred, label='Predicted Data', color='green')
plt.xlabel('Year')
plt.ylabel('IMR')
plt.title('IMR Prediction')
plt.legend()
plt.show()
```
